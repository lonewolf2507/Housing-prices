**Anova test is used when you have to compare multiple different group means all at the same time**

```python
import scipy.stats as stats
```

```python
stats.f_oneway(bang['Price'] , chen['Price'] , hyd['Price'] , mum['Price'] , delhi['Price'] , kolk['Price'])
```

The p-value is a probability that measures the evidence against the null hypothesis. Lower probabilities provide stronger evidence against the null hypothesis. A sufficiently large F-value indicates that the term or model is significant.

**In ANOVA, the null hypothesis is that there is no difference among group means**

As in our case the p value is low, which indicates that the null hypothesis we have is true
